```python
import pandas as pd
import numpy as np
import excel2json
import json

df_strength = pd.read_excel('data/xlsx/Strength AnalysisData.xlsx')
df_weakness = pd.read_excel("data/xlsx/Weakness Analysis Data.xlsx")
df_opportunity = pd.read_excel("data/xlsx/Opportunity Analysis Data.xlsx")
df_threat = pd.read_excel("data/xlsx/Threat Analysis Data.xlsx")

# combine
df_strength['Effect'] = 'POSITIVE'
df_weakness['Effect'] = 'NEGATIVE'
df_opportunity['Effect'] = 'POSITIVE'
df_threat['Effect'] = 'NEGATIVE'
df_positive = pd.concat([df_strength, df_opportunity])
df_negative = pd.concat([df_weakness, df_threat])

# Convert to JSON object
strength_json = df_strength.to_json(orient='records')
weakness_json = df_weakness.to_json(orient='records')
opportunity_json = df_opportunity.to_json(orient='records')
threat_json = df_threat.to_json(orient='records')
positive_json = df_positive.to_json(orient='records')
negative_json = df_negative.to_json(orient='records')
```

```python
# save json files
# strength
save_strength_json = open("strength_json.json", "w")
json.dump(strength_json, save_strength_json, indent = 6)
save_strength_json.close()
```

```python
# weakness
save_weakness_json = open("weakness_json.json", "w")
json.dump(weakness_json, save_weakness_json, indent = 6)
save_weakness_json.close()

# opportunity
save_opportunity_json = open("opportunity_json.json", "w")
json.dump(opportunity_json, save_opportunity_json, indent = 6)
save_opportunity_json.close()

# threat
save_threat_json = open("threat_json.json", "w")
json.dump(threat_json, save_threat_json, indent = 6)
save_threat_json.close()

# positive
save_positive_json = open("positive_json.json", "w")
json.dump(positive_json, save_positive_json, indent = 6)
save_positive_json.close()

# negative
save_negative_json = open("negative_json.json", "w")
json.dump(negative_json, save_negative_json, indent = 6)
save_negative_json.close()
```
